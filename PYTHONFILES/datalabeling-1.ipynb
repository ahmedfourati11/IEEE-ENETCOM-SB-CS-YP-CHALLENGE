```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

Liste_deepfake_edge=[]
Liste_deepfake_mask=[]
Liste_deepfake_image=[]
Label_deepfake=[]
import os
for dirname, _, filenames in os.walk('/kaggle/input/cstsyp2023/Dataset/Dataset/ICNC2023-Deepfakes (1)/edge'):
    for filename in filenames:
        Liste_deepfake_edge.append(os.path.join(dirname, filename))

for dirname, _, filenames in os.walk('/kaggle/input/cstsyp2023/Dataset/Dataset/ICNC2023-Deepfakes (1)/image'):
    for filename in filenames:
        Liste_deepfake_image.append(os.path.join(dirname, filename))

for dirname, _, filenames in os.walk('/kaggle/input/cstsyp2023/Dataset/Dataset/ICNC2023-Deepfakes (1)/mask'):
    for filename in filenames:
        Liste_deepfake_mask.append(os.path.join(dirname, filename))
# Pad list1 with None values to match the length of list2
Liste_deepfake_mask.extend([None for _ in range(len(Liste_deepfake_image) - len(Liste_deepfake_mask))])
Liste_deepfake_edge.extend([None for _ in range(len(Liste_deepfake_image) - len(Liste_deepfake_edge))])

lists = [Liste_deepfake_image,Liste_deepfake_mask,Liste_deepfake_edge]
    

most_nones = None
max_nones = 0

for list in lists:
    current_nones = 0
    for item in list:
        if item is None:
            current_nones += 1
    if current_nones > max_nones:
        max_nones = current_nones
        most_nones = list

for path in most_nones:
    if path==None:
        Label_deepfake.append(0)
    else:
        Label_deepfake.append(1)
        
print(len(Liste_deepfake_image))
print(len(Liste_deepfake_edge))
print(len(Liste_deepfake_mask))
print(len(Label_deepfake))
```

```python
import csv



# Get the lengths of the lists
lengths = [Liste_deepfake_image, Liste_deepfake_edge, Liste_deepfake_mask, Label_deepfake]

# Specify the CSV file path
csv_file_path = 'path_data_deepfake.csv'

# Write the lengths to the CSV file
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write header
    writer.writerow(['deepfake_image', 'deepfake_edge', 'deepfake_mask', 'deepfake'])
    
    # Write lengths
    writer.writerow(lengths)

print(f"Lengths written to {csv_file_path}")
```
